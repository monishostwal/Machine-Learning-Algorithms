```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('./Datasets/USA_Housing.csv')
```

```python
X=df.drop(['Price','Address'],axis=1)
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(X)
X=scaler.transform(X)
```

```python
y=df.iloc[:,5]
y=np.array(y)
y=y.reshape(5000,1)
```

```python
weights=np.random.uniform(-5,5,(5,1))
biases=np.random.uniform(-5,5,1)
```

```python
weights
```

```python
biases
```

```python
learning_rate=0.0001
epochs=100000
n=float(len(X))
```

```python
X.shape
```

```python
weights.shape
```

```python
for i in range(epochs):
    y_predicted=np.dot(X,weights)+biases
    delta=y_predicted-y
    loss=np.sum(delta**2)/2/n
    print(loss)
    delta_scaled=delta/n
    weights=weights-learning_rate*np.dot(X.T,delta_scaled)
    biases=biases-learning_rate*np.sum(delta_scaled)
```
